```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/oil-spills dataset/2015-2019.xlsx'
df5 = pd.read_excel(file_path)
mask = (df5['Recovered Spill Volume (bbl)'] > df5['Estimated Spill Volume (bbl)']) & ~df5['Recovered Spill Volume (bbl)'].isna()
df5.loc[mask, 'Recovered Spill Volume (bbl)'] = df5.loc[mask, 'Estimated Spill Volume (bbl)']

total_spilled_oil = df5['Estimated Spill Volume (bbl)'].sum()
total_recovered_oil = df5['Recovered Spill Volume (bbl)'].sum()
average_recovered_oil = total_recovered_oil / total_spilled_oil
df5['Recovered Spill Volume (bbl)'].fillna(df5['Estimated Spill Volume (bbl)'] * average_recovered_oil, inplace=True)
df5=df5.drop(['Column1','Column2','Column3','Column4'],axis=1)
df5.head(20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df['Date Reported'] = pd.to_datetime(df['Date Reported'])

df['Year'] = df['Date Reported'].dt.year

result_df = df.groupby('Year').agg({'Estimated Spill Volume (bbl)': 'sum', 'Recovered Spill Volume (bbl)': 'sum'}).reset_index()

print(result_df)
plt.figure(figsize=(10, 6))
plt.bar(result_df['Year'], result_df['Estimated Spill Volume (bbl)'], label='Estimated Spill Volume (bbl)')
plt.bar(result_df['Year'], result_df['Recovered Spill Volume (bbl)'], label='Recovered Spill Volume (bbl)', alpha=0.7)

plt.xlabel('Year')
plt.ylabel('Volume')
plt.title('Estimated Spill Volume and Recovered Estimated by Year')

plt.legend()

plt.show()
```
